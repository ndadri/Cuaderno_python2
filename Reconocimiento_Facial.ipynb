# Reconocimiento de imagenes OpenCV

```python
#libreria que me permite realizar el reconocimiento facial 
import cv2
#Importamos la libreria de numpy
import numpy as np
```

```python
# Cargar el harcascadefiel
face_cascade = cv2.CascadeClassifier('haar_cascade_files/haarcascade_frontalface_default.xml')
```

```python
#Comprobamos si el archivo cascade se a cargado correctamente
if face_cascade.empty():
    raise IOError('No se puede cargar correctamente el archivo')
```

```python
#Inicializamos el video de captura para el objeto
cap=cv2.VideoCapture(0)
```

```python
#Definir la escala
scalling_factor=0.5
```

```python
#Iteramos el video hasta presionar la tecla ESC
while True:
    #Capturamos el frame
    _, frame = cap.read()
    # Inicializamos el tamaño del frame
    frame = cv2.resize(frame, None,
                        fx=scalling_factor, fy=scalling_factor,
                        interpolation=cv2.INTER_AREA)
    #Convertimos la imagena escala de grises 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #Ejecutamos el detector de rostro en una esca de grises
    face_rects= face_cascade.detectMultiScale(gray, 1.3, 5)
    #Dibujar un rectangulo alrededor del rostro 
    for (x,y,w,h) in face_rects:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 3)
    #Mostramosla salida
    cv2.imshow('Detector de rostro', frame)

    #Vamos a detectar si el usuario presiono la tecla ESC 
    c = cv2.waitKey(1)
    if c==27:
        break
```

# Renderize el video capture

```python
cap.release()
```

# Cerramos la ventana

```python
cv2.destroyAllWindows()
```
