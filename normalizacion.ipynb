```python
from sklearn import preprocessing
import numpy as np
```

# Carga de datos de entrenamiento

- Se crea una matriz con 3 instancias (filas) y 3 atributos (columnas)

```python
X_train = np.array([[ 1., -4.,  5.],
                    [ 5.,  0.,  2. ],
                    [ 0.,  6., -3.]])
print("Datos de entrenamiento originales:")
print(X_train)
```

# Limpieza de datos: normalización (MinMaxScaler)

- El objetivo es transformar los atributos al rango [0, 1]

```python
normalizer = preprocessing.MinMaxScaler()
```

- 'fit_transform' calcula los mínimos/máximos y aplica la fórmula en un solo paso

```python
X_train_norm = normalizer.fit_transform(X_train)

print("\nDatos de entrenamiento normalizados:")
print(X_train_norm)
```

# Verificación de consistencia

- Comprobamos que aplicar 'fit' y luego 'transform' da el mismo resultado

```python
X_train_2 = np.copy(X_train)
f = normalizer.fit(X_train) # Aprende los parámetros (min/max)
X_train_norm_2 = f.transform(X_train) # Aplica la transformación

print("\n¿Los métodos son equivalentes?:", np.array_equiv(X_train_norm, X_train_norm_2))
```

# Normalizar un nuevo conjunto de datos (Test)

- **IMPORTANTE:** En test solo usamos 'transform'. No usamos 'fit' porque debemos usar la escala aprendida de los datos de entrenamiento.

```python
X_test = np.array([[ 1., -3.,  2.],
                    [ 3.,  0.,  0.],
                    [ 0.,  4., -1.]])
```

```python
X_test_norm = normalizer.transform(X_test)

print("\nNuevos datos de test normalizados (con la escala de entrenamiento):")
print(X_test_norm)
```
