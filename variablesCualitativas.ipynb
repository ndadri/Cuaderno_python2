```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creamos el DataFrame a partir de tu lista original 'nivel_estudios'
df = pd.DataFrame(nivel_estudios, columns=['Nivel'])
df
```

```python
# Calculamos las frecuencias de forma automática
frecuencias = df['Nivel'].value_counts()
```

```python
# 3. Mostrar la tabla de frecuencias (opcional pero útil)
print("Tabla de Frecuencias:")
print(frecuencias)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Diagrama de Barras con Pandas
frecuencias.plot(kind='bar', ax=ax1, color='skyblue', edgecolor='black')
ax1.set_title('Diagrama de Barras (Pandas)')
ax1.set_xlabel('Nivel de estudios')
ax1.set_ylabel('Frecuencia absoluta')
ax1.tick_params(axis='x', rotation=45)

# Diagrama de Sectores con Pandas
frecuencias.plot(kind='pie', ax=ax2, autopct='%1.1f%%', 
                 startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
ax2.set_title('Diagrama de Sectores (Pandas)')
ax2.set_ylabel('') 

plt.tight_layout()
plt.show()
```
